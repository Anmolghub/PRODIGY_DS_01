**Import Libraries**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv('/content/world_population.csv')

df.head()
```

**Dataset Overview**

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

**Check Missing Values**

```python
df.isnull().sum()
```

**Data Cleaning**

```python
df.duplicated().sum()
```

```python
# Remove if present
df.drop_duplicates(inplace=True)
```

__________________________
**Histogram — Population Distribution**

```python
plt.figure(figsize=(12,6))

sns.histplot(
    df['2022 Population'],
    bins=30,
    kde=True
)

plt.title('Distribution of World Population (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')

plt.show()
```

**Top 10 Most Populated Countries**

```python
top10 = df.sort_values(
    by='2022 Population',
    ascending=False
).head(10)
```

```python
top10[['Country/Territory','2022 Population']]
```

**Bar Chart**

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x='2022 Population',
    y='Country/Territory',
    data=top10
)

plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Country')

plt.show()
```

**Population by Continent**

```python
continent_pop = df.groupby(
    'Continent'
)['2022 Population'].sum().sort_values(
    ascending=False
)
```

```python
plt.figure(figsize=(10,6))

continent_pop.plot(
    kind='bar'
)

plt.title('Population by Continent (2022)')
plt.xlabel('Continent')
plt.ylabel('Population')

plt.show()
```

**Top 10 Densely Populated Countries**

```python
dense = df.sort_values(
    by='Density (per km²)',
    ascending=False
).head(10)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x='Density (per km²)',
    y='Country/Territory',
    data=dense
)

plt.title('Top 10 Most Densely Populated Countries')
plt.show()
```

**Population Growth Rate Analysis**

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df['Growth Rate'],
    bins=25,
    kde=True
)

plt.title('Distribution of Growth Rates')
plt.show()
```

**Correlation Analysis**

```python
numeric_df = df.select_dtypes(
    include=np.number
)

plt.figure(figsize=(14,10))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Matrix')
plt.show()
```
